```python
unique_courses_dict = {}

# Iterate through the grouped data and populate the dictionary
for course_id, course_edesc_list in df.groupby('COURSE_ID')['COURSE_EDESC'].unique().items():
    unique_courses_dict[course_id] = course_edesc_list[0]

print(unique_courses_dict)
```

```python
unique_courses_studyPlan = {}

# Iterate through the grouped data and populate the dictionary
for course_id, course_edesc_list in study_plan.groupby('COURSE_ID')['COURSE_DESCE'].unique().items():
    unique_courses_studyPlan[course_id] = course_edesc_list[0]

# Print the dictionary
print(unique_courses_studyPlan)
```

```python
# find keys present in unique_courses_dict but not in unique_courses_studyPlan
missing_keys_dict = set(unique_courses_dict.keys()) - set(unique_courses_studyPlan.keys())
missing_keys_studyPlan = set(unique_courses_studyPlan.keys()) - set(unique_courses_dict.keys())

# finding key-value pairs with different values for common keys to see if they align
different_values = {
    key: (unique_courses_dict[key], unique_courses_studyPlan[key])
    for key in set(unique_courses_dict.keys()) & set(unique_courses_studyPlan.keys())
    if unique_courses_dict[key] != unique_courses_studyPlan[key]
}
print("Keys with different values:")
for key, values in different_values.items():
    print(f"Key: {key}, Value in unique_courses_dict: {values[0]}, Value in unique_courses_studyPlan: {values[1]}")

print("\nKeys present in unique_courses_dict but not in unique_courses_studyPlan:")
for key in missing_keys_dict:
    print(f"Key: {key}, Value in unique_courses_dict: {unique_courses_dict[key]}")

print("\nKeys present in unique_courses_studyPlan but not in unique_courses_dict:")
for key in missing_keys_studyPlan:
    print(f"Key: {key}, Value in unique_courses_studyPlan: {unique_courses_studyPlan[key]}")
```
